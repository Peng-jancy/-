```python
# ===============================
# main_mnist.ipynb
# 入口文件：MNIST 手写字体识别
# ===============================

# 1. 导入模块
from data.mnist_data import load_data
from models.mnist_net import MNISTNet
from train.train_model import train_model
from yest.test_model import test_model
import matplotlib.pyplot as plt

# 2. 加载数据
train_loader, test_loader = load_data()

# 3. 初始化模型
model = MNISTNet()

# 4. 训练模型（返回模型和 loss_list）
model, loss_list = train_model(model, train_loader, num_epochs=5, lr=0.001)

# 5. 测试模型
test_model(model, test_loader)

# 6. 绘制 Loss 曲线
plt.plot(range(1, len(loss_list)+1), loss_list, marker='o')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("MNIST Training Loss Curve")
plt.grid(True)
plt.show()
```
